```python
import re
import sys
import numpy as np
import pandas as pd
from glob import glob
sys.path.append("../src")
from paper_reader import Paper
```

```python
papersList = [paperPath.split('/')[-1] for paperPath in glob("../papers/*.pdf")]
```

```python
papersList
```

```python
details = []
for paperName in papersList:
    paper = Paper(paperName)
    
    details.append({
        "pdf": paperName,
        "doi": paper.doi,
        'title': paper.title,
        'authors': ";".join(paper.author),
        'journal': paper.journal,
        'publisher': paper.publisher,
        'year': paper.year,
        'keywords': ";".join(paper.keywords)
    })
data = pd.DataFrame(details)
```

```python
filterData = data[(data['doi']!='') & (data['keywords']!="")]
filterData.loc[:, ["SC"]] = "Computer Science"
filterData.loc[:, ["ID"]] = "generic keyword1;keyword2"
```
